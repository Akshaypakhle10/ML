##Web Scraping the Transcripts 

Source - https://bigbangtrans.wordpress.com

```
import requests
from bs4 import BeautifulSoup
r = requests.get('https://bigbangtrans.wordpress.com')
soup = BeautifulSoup(r.text)
```

```
def get_episode_links(soup):
    episode_links = []
    for link in soup.find("div", {"id": "pages-2"}).find_all('a'):
        txt = link.get('href')
        if 'series' in txt:
            episode_links.append(txt)
    return episode_links

def get_full_episode_transcript(soup):
    txt_spans = soup.find("div", {"class": "entrytext"}).find_all('span')
    txt = ""
    for txt_span in txt_spans:
        string = txt_span.string
        if string == 'Written by Chuck Lorre and Bill Prady':
            break
        if string is not None:
            txt = txt + string + "\n"
    return txt

def get_episode_title(soup):
    return str(soup.find("h2", {"class": "title"}).string)

def get_json_of_all_episodes(soup):
    episode_transcripts = {}
    episode_links = get_episode_links(soup)
    for episode_link in episode_links:
        r = requests.get(episode_link)
        soup = BeautifulSoup(r.text)
        title = get_episode_title(soup)
        transcript = get_full_episode_transcript(soup)
        episode_transcripts[title] = transcript
    return episode_transcripts

def get_personas_from_episodes(soup):
    personas = {}
    episode_links = get_episode_links(soup)
    for episode_link in episode_links:
        r = requests.get(episode_link)
        soup = BeautifulSoup(r.text)
        transcript = get_full_episode_transcript(soup)
        for line in transcript.strip().split("\n"):
            #print(line)
            split_line = line.split(":")
            #print(split_line)
            if len(split_line) == 2:
                if split_line[0] in personas:
                    personas[split_line[0]] = personas[split_line[0]]+[split_line[1]]
                else:
                    personas[split_line[0]] = [split_line[1]]
    return personas

def save_json_to_file(json_file, save_path):
    import json
    with open(save_path, 'w') as file:
        file.write(json.dumps(json_file))
    print("File saved at: ", save_path)


        
```

```
all_episode_transcripts = get_json_of_all_episodes(soup)
all_episode_personas = get_personas_from_episodes(soup)
save_path_transcripts = '/content/drive/My Drive/datasets/BigBangTheory/full_transcripts.json'
save_path_personas = '/content/drive/My Drive/datasets/BigBangTheory/full_personas.json'
save_json_to_file(all_episode_transcripts, save_path_transcripts)
save_json_to_file(all_episode_personas, save_path_personas)
```
